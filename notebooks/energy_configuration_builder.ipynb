```python
import json
```

```python
# initialize the dictionary
configuration = {
    "method": "optimize",           # request a geometry optimization
    "options": "",                  # parameters specific to the optimizer that you will use
    "atoms": "",                    # location of .npz file containing a list of lists of atomic symbols
    "coordinates": "",              # location of .npz file containing a list of lists of atomic coordinates
    "charge": 0,                    # an int or list of ints for the charge on each structure
    "spin": 1,                      # an int or list of ints for the spin multiplicity of each structure
    "output_dir": "."               # the location where you expect .npz zipped output files to land
}
```

```python
optimizer_options = {
    # these first three parameters are `shared` between the optimizers
    "optimizer": "ase",                                             # the optimizer that you would like to use.
    "logging": "",                                                  # the path to where you expect stdout/stderr to be sent. if empty string -> goes to console.
    "output": ["coordinates", "atoms", "energies", "gradients"],    # the flavors of information that you would like returned to you under `output_dir`
    # the rest of these are specific to the ase optimizer
    "fmax": 0.02,                                                   # termination condition: the maximum magnitude of force that any individual atom can have 
    "steps": 500,                                                   # termination condition: maximum number of steps taken during optimization 
}
```

```python
configuration["options"] = optimizer_options
```

```python
configuration_path = "./config.json"
```

```python
with open(configuration_path, "w") as f:
    json.dump(configuration, configuration_path)
```
