```python
import json
```

```python
# initialize the dictionary
config = {
    "method": "energy",     # request an energy computation
    "charge": 0,            # an int or list of ints for the charge on each structure
    "spin": 1,              # an int or list of ints for the spin multiplicity of each structure
    "output_dir": "."       # the location where you expect .npz zipped output files to land
}
```

```python
# the server instance of mlipenv can handle either a folder containing .xyz file(s),
path_to_xyz_folder = "..."
config["structures"] = path_to_xyz_folder
```

```python
# or a pair of paths to .npz files with all of the coordinates and atoms aligned and in your desired computation order.
config["atoms"] = "mlipenv/test/atoms.npz"              # location of .npz file containing a list of lists of atomic symbols
config["coordinates"] = "mlipenv/test/coordinates.npz",  # location of .npz file containing a list of lists of atomic coordinates
```

```python
# parameters for the energy computation.
energy_options = {
    "order": 1    # example: when order=1, returns energy (0-tensor) and gradients (1-tensor). return each nth-derivative tensor of the energy, for n <= order.
}

config["energy_options"] = energy_options
```

```python
config_write_path = "./energy_config.json"
```

```python
with open(config_write_path, "w") as f:
    json.dump(config, f)
```
